```python
"""


Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is
not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are 
pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

"""
```

```python
def P(n):
    return int(n * (3 * n - 1) / 2)
```

```python
for n in range(10):
    print(P(n), end=', ')
```

```python
many_pentagon_numbers = []
for n in range(1, 10001):
    many_pentagon_numbers += [P(n)]
```

```python
for i in range(10000):
    num1 = many_pentagon_numbers[i]
    for j in range(10000):
        num2 = many_pentagon_numbers[j]
        
        sum_penta = num1 + num2
        dif_penta = abs(num1 - num2)
        
        if sum_penta in many_pentagon_numbers and dif_penta in many_pentagon_numbers:
            print(num1, num2, sum_penta, dif_penta)
```
