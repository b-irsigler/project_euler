```python
"""


The sequence of triangle numbers is generated by adding the natural numbers. So the 7th 
triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

"""
```

```python
import numpy as np
```

```python
def triangle_number(n):
    return np.arange(1, n + 1).sum()
```

```python
for i in range(1, 8):
    print(i, triangle_number(i))
```

```python
def factors(number):
    fac = []
    for i in range(1, number + 1):
        if number % i == 0:
            fac += [i]
    return fac
```

```python
for i in range(1, 8):
    tri = triangle_number(i)
    print(tri, factors(tri))
```

```python
max_tri_facs = 0
for i in range(10000):
    if i % 100 == 0: print(i, end = '\r')
    tri = triangle_number(i)
    tri_facs = len(factors(tri))
    if tri_facs > max_tri_facs:
        max_tri_facs = tri_facs
        print(tri, tri_facs)
        if tri_facs > 500:
            break
```
